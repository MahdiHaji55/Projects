Many investment instruments provided by banks, such as Guaranteed Investment Certificates (GICs) pay interest not only on the initial investment (the *principal*) but also on the interest that accumulates over time (*compound* interest).

Investment advisors normally discuss options with the clients by focusing on the two following points:

1. **Purchase a GIC at a specific interest rate**; and then
1. **Let their GIC earn compound interest for a certain number of years.**

The formula for calculating the balance of such an investment after a specified number of years is as follows:

$A = P (1 + \frac{r}{n})^{nt}$

where:

* $A$ is the amount of money in the account after the specific number of years.
* $P$ is the amount deposited into the account.
* $r$ is the **annual** interest rate.
* $n$ is the number of times per year that the interest is compounded
* $t$ is the specific number of years.

Assuming that $n = 12$ (as this would represent an investment compounded monthly) then $P$ would be the amount contributed to the investment each month.

You are to apply the above formula in order to compute the value of an RESP (Registered Education Savings Plan) for little Tanya assuming her parents invest in GICs as described above. An RESP is used to defer tax on interest, and hence is attractive for those trying to save for their children's post-secondary education.

Given the age of the child at the start of the RESP, the monthly amount to be saved, and the annual rate of return of the investment, your program is to calculate the total amount of the RESP when the child reaches the age of 18. More precisely, you are now print a table showing the value of the RESP at the end of each year for which funds were saved in the RESP. That is, after asking for:

````python
# Put in solution to Part B here.
# Use only this code cell.
AGE = int(input("The age of the child at the start of the RESP:"))
AMOUNT = int(input("The amount to be invested each month:"))
RATE = int(input("The annual rate of return for the investment, in percent:"))/100
N = 12

print("age""    ""RESP value")

for i in range (AGE+1,19,1):
    a = RATE/N
    b = 1 + a
    c = N*i
    d = b**c
    e = AMOUNT*d
    print(i,"      $",format(e,"0.2f"))









    


#There is a minor problem(Formula and the way question is presented and needs specificly) with this part of the assigment (PART B) so instead i used the formula given above and used the key concept we learned in class to write the best possible code i could write and that is appropriate to the cource material




````
