````python
STOCKDATA_LIST_01=["2021,06,01,27.33,27.50,27.20,27.43,26.28,2344000",
    "2021,06,02,27.48,27.60,27.42,27.51,26.35,1531300",
    "2021,06,03,27.52,27.75,27.50,27.63,26.47,2795700",
    "2021,06,04,27.72,27.89,27.66,27.88,26.71,1180100",
    "2021,06,07,27.96,28.11,27.93,28.07,26.89,2597200",
    "2021,06,08,28.11,28.14,27.85,28.00,26.82,4005900",
    "2021,06,09,27.53,27.68,27.36,27.49,26.63,5802000",
    "2021,06,10,27.46,27.57,27.33,27.42,26.57,3207400",
    "2021,06,11,27.44,27.59,27.40,27.49,26.63,2310500",
    "2021,06,14,27.49,27.64,27.39,27.64,26.78,1217600",
    "2021,06,15,27.70,27.91,27.68,27.80,26.94,1687700",
    "2021,06,16,27.75,27.88,27.72,27.72,26.86,1731200",
    "2021,06,17,27.73,27.92,27.72,27.89,27.02,1215800",
    "2021,06,18,27.76,27.88,27.61,27.64,26.78,4981800",
    "2021,06,21,27.65,27.78,27.58,27.73,26.87,3472700",
    "2021,06,22,27.79,27.81,27.50,27.80,26.94,2025500",
    "2021,06,23,27.75,27.75,27.52,27.58,26.72,1201300",
    "2021,06,24,27.65,27.68,27.40,27.54,26.68,1424800",
    "2021,06,25,27.41,27.52,27.33,27.50,26.64,965900",
    "2021,06,28,27.54,27.77,27.47,27.69,26.83,2190200",
    "2021,06,29,27.71,28.04,27.71,27.88,27.01,2461300",
    "2021,06,30,27.90,27.90,27.58,27.80,26.94,1642300"
]

STOCKDATA_LIST_02=["2021,12,01,96.57,97.05,95.47,95.58,94.93,361500",
    "2021,12,02,95.53,98.57,95.28,97.79,97.12,547400",
    "2021,12,03,97.65,98.33,96.91,97.39,96.73,274500",
    "2021,12,06,97.52,98.64,97.39,98.24,97.57,257700",
    "2021,12,07,98.33,98.75,97.79,98.04,97.37,421600",
    "2021,12,08,97.70,97.92,96.57,97.05,96.39,677500",
    "2021,12,09,97.15,98.87,97.14,98.47,97.80,606700",
    "2021,12,10,98.63,99.77,98.42,99.21,98.53,563500",
    "2021,12,13,98.75,100.78,98.62,100.05,99.37,745200",
    "2021,12,14,99.02,101.03,98.97,100.86,100.54,449500",
    "2021,12,15,101.00,103.15,100.70,102.53,102.20,833300",
    "2021,12,16,102.37,104.08,102.27,102.96,102.63,542200",
    "2021,12,17,102.96,104.85,102.57,103.54,103.21,1335200",
    "2021,12,20,103.41,104.25,102.83,102.99,102.66,529100",
    "2021,12,21,103.14,104.18,102.80,103.33,103.00,396100",
    "2021,12,22,103.14,103.79,102.67,103.75,103.42,320500",
    "2021,12,23,103.46,103.92,103.08,103.18,102.85,466700",
    "2021,12,24,103.48,103.82,103.15,103.58,103.25,112700",
    "2021,12,29,103.60,105.15,103.60,104.42,104.09,349200",
    "2021,12,30,104.44,104.82,103.66,103.78,103.45,126000",
    "2021,12,31,103.65,103.95,103.06,103.64,103.31,178700"
]


def compute_month_stats(data):
    """
    function: compute_month_stats
    
    Given a list of strings containing stock-trading data for one month,
    compute summary statistics for that month (year, month,
    highest price in month, lowest price in month, total volume
    for the month, average daily volume rounded to the nearest integer,
    and the number of days where the opening price greater than
    the closing price) and return all this as a tuple of the form
    (int, int, float, float, int, int, int).
       
    Input parameter:
    ----------------
    
    * data: A list of strings where each string contains one day's
    trading information. All of the data in the list is for a
    single month. The values are separated by commas, and each day's
    list has the same number of values. The order of these values
    is: year, month, day, open price, highest price, lowest price,
    closing price, adjusted-closing price, total volume
      
    Output:
    -------
    *  There is no output to the console from this function.
    
    Return value:
    -------------
    A tuple with seven values is returned. The values are: the year;
    the month; the highest price the stock traded in the month;
    the lowest price the stock traded in the month; the total volume
    of shares traded in the month; the average daily volume of shares
    traded in the month, rounded to the nearest integer; and the number
    of days where the opening price is higher than the closing price.
    Note that in this function you will ignore the adjusted-close values.
    """

    
    
    high_month = []
    low_month = []
    total_month = []
    days_openclose = []
    
    for day_data in data:
        day_splited = day_data.split(",")
        
        highest_day = day_splited[4]
        high_month.append(float(highest_day))
        
        lowest_day = day_splited[5]
        low_month.append(float(lowest_day))
        
        volume = day_splited[8]
        total_month.append(int(volume))
        
        if day_splited[3] > day_splited[6]:
            days_openclose.append(int(day_data[3]))
                  
    num_days = len(days_openclose)
    t_volume = sum(total_month)
    average = t_volume / len(data)
    min_low = min(low_month)
    max_high = max(high_month)
    
    summary_stats = (int(day_splited[0]), int(day_splited[1]), float(max_high), float(min_low), int(t_volume), int(average), int(num_days))
            
                  
    
    # At present, the function return a bunch of zeros. You will,
    # of course, change this line as a result of your work completing
    # this part of the assignment.
    
    return (summary_stats) 
    

# Do not modify this main() function.
def main():
    stock_data = [STOCKDATA_LIST_01, STOCKDATA_LIST_02]
    
    for month_data in stock_data:
        stats = compute_month_stats(month_data)
        for val in stats[:-1]:
            print(val, ",", sep="", end="")
        print(stats[-1])
    
main()
````
