### Writing a class: `SimpleCipher`

Consider a string consisting of all lowercase letters plus a space, called a **plaintext alphabet**:
```
"abcdefghijklmnopqrstuvwxyz "
```
and consider another string with exactly the same letters but shuffled randomly which we call a **ciphertext alphabet**, for example:
```
"izg jcdlfxomvkwqnyateuhpbsr"
```

A **simple substitution cipher** is one which takes each letter of a **plaintext** message and converts it into its **ciphertext**. For example, the plaintext message:
```
"hello"
```
becomes the following as ciphertext:
```
"ljmmw"
```

If we place the alphabets one above the other, you can see how the `h` becomes `l`, `e` becomes `j`, `l` becomes `m`, and `o` becomes `w`.

```
plaintext alphabet:  "abcdefghijklmnopqrstuvwxyz "
ciphertext alphabet: "izg jcdlfxomvkwqnyateuhpbsr"
```

You are to write a class named `SimpleCipher` that:
* Takes a plaintext alphabet and a ciphertext alphabet as arguments in the constructor.
* Implements a method called `encipher` which takes a plaintext string and returns its ciphertext equivalent.
* Implements a method called `decipher` which takes a ciphertext string and returns its plaintext equivalent.

In the code cell provided to you, there is a `main` function which creates several different instances of `SimpleCipher` and uses them to encipher and decipher a few different strings.

Please note that **all** of the code you must write for this part must be located with the class (i.e. no functions outside of the class are permitted).

````python
# This class is clearly incomplete.

# You may assume that plaintext and ciphertext messages 
# will contain only lower-case letters and spaces (i.e. no numbers
# or other punctuation will appear).

class SimpleCipher:
    
    def __init__(self, p_alphabet, c_alphabet):
        self.p_alphabet = p_alphabet
        self.c_alphabet = c_alphabet
        return         #returns the data as variables assigned from the lists in main to the functions within the class for use  
    
    
    
    
    def encipher(self, plaintext):
        
        self.plaintext = plaintext
        codebook = {}
        for i, plain in enumerate (self.p_alphabet):
            codebook[plain] = self.c_alphabet[i]             #makes a codebook that plain alphabet is assigned as key and cipher assigned as value within the dictionary 
        
        cipher_letters = []
        for letters in self.plaintext:                       #goes through the plain text provided in the main 
            for item in codebook:                            #goes through the dictionary created above,
                if item == letters:                          #then check if there are matching words,
                    cipher_letters.append(codebook[item])    #if there are matching words it adds it to a list called cipher_letters
        
        
        ciphered_word = "".join(cipher_letters)              #it combines our list together as one word 
    
        
        return ciphered_word                                 #then returns the ciphered word
    
    
    
    
    def decipher(self, ciphertext):                          #this function fucntions the same way but it does the oppsite which is, 
                                                             #getting ciphered word and make a dictionary then look through the dictionary to,
        self.ciphertext = ciphertext                         #find the matching letter and decipher the word given
        decodebook = {}
        for i, cipher in enumerate (self.c_alphabet):
            decodebook[cipher] = self.p_alphabet[i]
            
        plain_letters = []
        for letters in self.ciphertext:
            for item in decodebook:
                if item == letters:
                    plain_letters.append(decodebook[item])
                    
        deciphered_word = "".join(plain_letters)
        
        
        return deciphered_word                                      
    
    
    
    
    def get_ciphertext_alphabet(self):
        
        return self.c_alphabet                              #simple getter of ciphertext
    

# Do not modify this main() function.
#

def main():
    plaintext_alphabet   = "abcdefghijklmnopqrstuvwxyz "
    
    ciphertext_alphabet = ["izg jcdlfxomvkwqnyateuhpbsr",
                        "gtyvibpjxska hucrdqnlmwfzoe",
                        "uplcjzqn kdmrvhsfoyiteawxgb"]
    
    plaintext = ["hello", "shall i compare thee to a summers day",
                 "the rain in spain falls mainly on the plain"]
    
    ciphertext = [ ["ljmmw", "alimmrfrgwvqiyjrtljjrtwriraevvjyar ib",
                    "tljryifkrfkraqifkrcimmarvifkmbrwkrtljrqmifk"],
                   ["jiaau", "qjgaaexeyu cgdienjiienuegeql  idqevgz",
                    "njiedgxhexheqcgxhebgaaqe gxhazeuhenjiecagxh"],
                   ["njmmh", "ynummb blhrsuojbinjjbihbubytrrjoybcux",
                    "injbou vb vbysu vbzummybru vmxbhvbinjbsmu v"]
                 ]

    simple_ciphers = []
    for ca in ciphertext_alphabet:
        cipher = SimpleCipher(plaintext_alphabet, ca)
        simple_ciphers.append(cipher)
        
    # Simple sanity check (encipher)
    print("Sanity check: Does encipher example from description work?")
    print(repr("hello"), "-->", repr(ciphertext[0][0]),
          simple_ciphers[0].encipher("hello") == ciphertext [0][0])
    print()
    
    # Longer sanity-check (encipher)
    print("Sanity check: Does a longer encipher example work?")
    print(repr(plaintext[1]), "-->", repr(ciphertext[2][1]),
          simple_ciphers[2].encipher(plaintext[1]) == ciphertext [2][1])
    print()
    
    # Longer sanity-check (decipher)
    print("Sanity check: Does a longer decipher example work?")
    print(repr(ciphertext[1][2]), "-->", repr(plaintext[2]),
          simple_ciphers[1].decipher(ciphertext[1][2]) == plaintext[2])
    print()
    
    # Getter working?
    print("Sanity check: getter for ciphertext alphabet working?",
          simple_ciphers[1].get_ciphertext_alphabet() == ciphertext_alphabet[1])
    
main()
````
