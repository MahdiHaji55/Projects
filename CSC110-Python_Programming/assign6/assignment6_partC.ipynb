Your task in this last part of the assignment is to analyze the data on food-consumption behavior. **Several `CSV` files are provided with this assignment** that contain data regarding which organisms eat which foods (with "foods" here possibly other organisms). All provided data files use a comma-separated format (*i.e.*, what we expect within a `CSV` file). Each line in the file contains data about the eating habits of an organism and is provided in this format:

```<eater>,<eaten 1>,<eaten 2>, ...,<eaten n>```

`zebra,grass`

which means that `zebra`s eat only grass.

#### Functions 1 & 2:  `read_food_file(filename)` and `dump_data(eat_data)`

The first function `read_food_file` will open the file corresponding to the filename provided and create a dictionary of lists. That is, each entry in the dictionary will have as its key an eater (*i.e.*, first item in a line) and as a value all of the things which that eater will eat (i.e., second and following items in a line).

All items in a line are separated by commas (as we would expected with a `CSV` file) and you can assume all files provided are in the correct format (*i.e.*, you do not need to write error-checking code for input). However, you should print out an error message if the file corresponding to the filename doesnâ€™t exist, afterwards returning an empty dictionary.

The second function `dump_data` will print a readable version of the data constructed by the first function. Also note the eaters are printed in sorted order (i.e., alphabetical), and within each line the eaten are also printed in sorted order. #### Function 3:  `compute_only_eaters(eat_data)`

This third function uses the dictionary constructed by `read_food_file` and computes the eaters *who are also not eaten*. Once this is computed, these eaters are returned in a list in alphabetical order. For example, the output expected from this function given `food01.csv` is used by the `main` function to produce this output line:

#### Function 4:  `compute_only_eaten(eat_data)`

This fourth function uses the dictionary constructed by `read_food_file` and computes all which is eaten but which is not also an eater. (Hint: Something which is eaten but is not an eater will have no key in the dictionary.) Once computed the only-eaten items are returned in a list in alphabetical order. For example, the output expected from this function given `food01.csv` is used by the `main` function to produce this output line:

#### Function 5:  `compute_fussiest_eaters(eat_data)`

This fifth function uses the dictionary constructed by `read_food_file` and looks again at all of the eaters. However, the function will determine the eaters having the *fewest number of items eaten*. (Hint: You will want to compute the minimum of the eaten-item list lengths.) Once this is computed, the fussiest eaters are returned in a list in alphabetical order. #### Function 5:  `compute_most_delicious(eat_data)`

This last function uses the dictionary constructed by `read_food_file` and looks again at all of the eaten. However, the function will determine which, of those that are eaten, *have the largest number of eaters*. (Hint: You may want to compute a local dictionary where each of the eaten is a key and the corresponding value is a list of those that eat the eaten.) Once computed, the most delicious of the eaten is returned in a list in alphabetical order.
````python
def read_food_file(filename):
    
    file_read = open(filename, "r")    
    
    
    result = {}
    for row in file_read:                       
        rowstrip = row.strip()
        rowsplit = rowstrip.split(",")          
        in_order_split = rowsplit[1:]           
        in_order_split.sort()                   
        result[rowsplit[0]] = in_order_split    
        
        
    result = dict(sorted(result.items()))       
        
    
        
    return result                               


def dump_data(eat_data):
    
    for data in eat_data:
        print(data, ":", ",".join(eat_data[data]))     
        
           
    

def compute_only_eaters(life):
    
    list_of_not_eaten = []
    list_of_all_eatens = []
    for key in life:
        list_of_all_eatens.extend(life[key])
    for key in life:
        if key not in list_of_all_eatens:
            list_of_not_eaten.append(key)

    return sorted(list_of_not_eaten)
    


def compute_only_eaten(life):
    
    list_of_all_eatens = []
    for key in life:
        list_of_all_eatens.extend(life[key])

    no_duplicate_list_all_eatens = list(dict.fromkeys(list_of_all_eatens))

    for key in life:
        if key in list_of_all_eatens:
            no_duplicate_list_all_eatens.remove(key)

    return sorted(no_duplicate_list_all_eatens)


def compute_fussiest_eaters(life):
    
    list_of_fussiest_eaters = []
    dict_of_numbers = {}
    for key in life:
        dict_of_numbers[key] = len(life[key])

    minimum_value = min(dict_of_numbers.values())

    for key in dict_of_numbers:
        if dict_of_numbers[key] == minimum_value:
            list_of_fussiest_eaters.append(key)

    return sorted(list_of_fussiest_eaters)


def compute_most_delicious(life):
    
    inverted_dict = {}

    for key in life:
        for eaten in life[key]:
            inverted_dict[eaten] = []
    for key in life:
        for eaten in life[key]:
            inverted_dict[eaten].append(key)

    dict_of_numbers = {}
    for key in inverted_dict:
        dict_of_numbers[key] = len(inverted_dict[key])

    maximum_value = max(dict_of_numbers.values())

    list_of_most_delicious = []
    for key in dict_of_numbers:
        if dict_of_numbers[key] == maximum_value:
            list_of_most_delicious.append(key)

    return sorted(list_of_most_delicious)




def main():
    filename = input("Name of CSV file? ")
    life = read_food_file(filename)
    if life == {}: 
        return

    li = dump_data(life)
    print()

    li = compute_only_eaters(life)
    print("Only eaters:", ", ".join(li))
    print()

    li = compute_only_eaten(life)
    print("Only eaten:", ", ".join(li))
    print()

    li = compute_fussiest_eaters(life)
    print("Fussiest eaters:", ", ".join(li))
    print()

    li = compute_most_delicious(life)
    print("Tastiest entry:", ", ".join(li))


main()
````
