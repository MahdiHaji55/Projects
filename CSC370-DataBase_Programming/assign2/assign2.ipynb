Consider the relation $R$ with four attributed $ABCD$.

For each of the two following sets of FDs, determine whether or not the $R$ would be in BCNF. If the FDs set indicates that $R$ is not in BCNF, then decompose into BCNF. Prove the result (either the decomposed $R$ or the original) for each case produces a lossless join.

1. $A \rightarrow B$, $BC \rightarrow D$, $A \rightarrow C$
2. $AB \rightarrow C$, $AB \rightarrow D$, $C \rightarrow A$, $D \rightarrow B$

#### Answers for A
[Instructor only: A mark --]

*Your answers for Question A begin here*

**1.**

**Checking key:** $A$ is a candidate key since $A$ determines $B$, $C$, and indirectly $D$ via $BC$ $\rightarrow$ $D$.

**Checking BNCF:** $A \rightarrow B$ and $A \rightarrow C$ comply with BNCF as $A$ is a superkey, also $BC \rightarrow D$ does not comply since $BC$ is not a superkey from A is a candidate key.

**Decompostion:** Decompose $R$ into $R1(BCD)$ and $R2(ABC)$. $R1$ HANDLES $BC \rightarrow D$, ensuring BCNF as $BC$ acts as a superkey there and $R2$ will key the rest, so it will still be in BCNF.

**Checking loseless join:** Decomposition is loseless since $BC$, which is the common attribute set between $R1$ and $R2$, forms a candidate key is $R1$.

**Conclusion:** The original $R$ is not in BCNF due to $BC \rightarrow D$, and by decomposing into $R1(BCD)$ AND $R2(ABC)$, we achieve BCNF while maintaining a lossless join, ensuring data integrity.

**2.**

**Checking key:** Both $AB$ AND $CD$ can be considered candidate keys due to the mutual dependencies among $A$, $B$, $C$, and, $D$.

**Checking BCNF:** All functional dependencies satisfy BCNF criteria since every attribute is part of a superkey due to the circular dependencies ($AB$ for $C$ and $D$, $C$ for $A$, $D$ for $B$).

**Decomposition:** No decomposition is needed because $R$ already satisfies BCNF. Every FD's left side is a superkey or part of a superkey.

**Checking lossless join:** Not applicable as $R$ was not decomposed; it inherently ensures a lossless join due to its structure.

**Conclusion:** $R$ is in BCNF as is, with no need for decomposition, maintaining data integrity and lossless join capability.
